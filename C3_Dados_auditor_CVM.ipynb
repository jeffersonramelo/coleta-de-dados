```python
import pandas as pd
import wget
from zipfile import ZipFile
```

```python
#Acessando a base de dados e criando arquivos históricos
url_base = 'https://dados.cvm.gov.br/dados/CIA_ABERTA/DOC/FRE/DADOS/'
```

```python
arquivos_zip = []
for ano in range(2010, 2024):
  arquivos_zip.append(f'fre_cia_aberta_{ano}.zip')
```

```python
arquivos_zip
```

```python
for arq in arquivos_zip:
  wget.download(url_base+arq)
```

```python
#extraindo os arquivos do ZIP
for arq in arquivos_zip:
  ZipFile(arq, 'r').extractall('CVM')
```

Consolidar todos os arquivos CSV em um só CSV. exemplo do csv: fre_cia_auditor: Auditores Independentes - Identificação e Remuneração (Itens 2.1/2 do Anexo 24 da ICVM 480)

```python
import os
import pandas as pd

# Diretório dos arquivos CSV
diretorio = '/content/CVM/'

# Crie uma lista vazia para armazenar os DataFrames
dataframes = []

# Crie um loop para percorrer cada ano de 2010 a 2023
for ano in range(2010, 2024):
    # Construa o caminho completo para o arquivo CSV atual
    nome_arquivo = f"fre_cia_aberta_auditor_{ano}.csv"
    caminho_arquivo = os.path.join(diretorio, nome_arquivo)

    # Carregue o arquivo CSV atual com a codificação 'latin-1' e o delimitador ';'
    df = pd.read_csv(caminho_arquivo, encoding='latin-1', delimiter=';')

    # Adicione o DataFrame à lista
    dataframes.append(df)

# Concatene os DataFrames em um único DataFrame
df_final = pd.concat(dataframes)

# Salve o DataFrame final em um novo arquivo CSV
caminho_saida = os.path.join(diretorio, "fre_cia_aberta_auditor_todos.csv")
df_final.to_csv(caminho_saida, index=False)
```

```python
import pandas as pd

# Lendo o arquivo CSV usando o pandas
df = pd.read_csv('/content/CVM/fre_cia_aberta_auditor_todos.csv')
```

```python
df
```

```python
print(df.loc[0, 'Remuneracao_Auditor'])
```

```python
import re

# Expressão regular para encontrar um valor monetário
pattern = r'R\$\s?\d{1,3}(?:\.\d{3})*(?:,\d{2}|\d{3}\s?milhões)'

# Criar uma lista para armazenar os valores extraídos
valores = []

# Iterar sobre a coluna 'Remuneracao_Auditor' e extrair os valores
for text in df['Remuneracao_Auditor']:
    text = str(text)  # Converter para string
    match = re.search(pattern, text)
    if match:
        valor = match.group()
    else:
        valor = ''
    valores.append(valor)

# Adicionar os valores extraídos como uma nova coluna no DataFrame
df['Valores_Extraidos'] = valores
```

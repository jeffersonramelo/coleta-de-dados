Filtro por empresa e por contas

```
import pandas as pd
from IPython.display import display

# Supondo que BPA é o DataFrame original

# Seleciona as colunas específicas e remove duplicatas
empresas = BPA[['DENOM_CIA', 'DT_REFER', 'DS_CONTA', 'CD_CVM', 'VL_CONTA']].drop_duplicates()

# Filtra as linhas onde 'DS_CONTA' é igual a 'Ativo Total' ou 'Caixa e Equivalentes de Caixa'
contas_desejadas = ['Ativo Total', 'Caixa e Equivalentes de Caixa']
ativo_total_caixa = empresas[empresas['DS_CONTA'].isin(contas_desejadas)]

# Filtra as linhas onde 'DENOM_CIA' é igual a 'MAGAZINE LUIZA S.A.'
companhia = ativo_total_caixa[ativo_total_caixa['DENOM_CIA'] == 'MAGAZINE LUIZA S.A.']

# Pivot para criar colunas separadas para 'Ativo Total' e 'Caixa e Equivalentes de Caixa'
pivot_companhia = companhia.pivot_table(index=['DT_REFER', 'DENOM_CIA', 'CD_CVM'],
                                        columns='DS_CONTA',
                                        values='VL_CONTA',
                                        aggfunc='last').reset_index()

# Renomeia as colunas para serem mais claras
pivot_companhia.columns.name = None
pivot_companhia.rename(columns={
    'Ativo Total': 'Ativo_Total',
    'Caixa e Equivalentes de Caixa': 'Caixa_Equivalentes_Caixa'
}, inplace=True)

# Exibe a tabela pivotada
display(pivot_companhia)
```

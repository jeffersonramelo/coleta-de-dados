Filtro por conta

```
import pandas as pd
from IPython.display import display

# Supondo que BPA é o DataFrame original

# Seleciona as colunas específicas e remove duplicatas
empresas = BPA[['DENOM_CIA', 'DT_REFER', 'DS_CONTA', 'CD_CVM', 'VL_CONTA']].drop_duplicates()

# Lista de contas desejadas
contas_desejadas = [
    'Ativo Circulante', 'Ativo Realizável a Longo Prazo', 'Estoques',
    'Caixa e Equivalentes de Caixa', 'Ativo Total', 'Imobilizado',
    'Intangível', 'Contas a Receber'
]

# Filtra as linhas onde 'DS_CONTA' é igual a uma das contas desejadas
filtradas = empresas[empresas['DS_CONTA'].isin(contas_desejadas)]

# Pivot para criar colunas separadas para cada conta
pivot_filtradas = filtradas.pivot_table(index=['DT_REFER', 'DENOM_CIA', 'CD_CVM'],
                                        columns='DS_CONTA',
                                        values='VL_CONTA',
                                        aggfunc='last').reset_index()

# Renomeia as colunas para remover a hierarquia de colunas
pivot_filtradas.columns.name = None

# Renomeia as colunas para serem mais claras
pivot_filtradas.rename(columns={
    'Ativo Circulante': 'Ativo_Circulante',
    'Ativo Realizável a Longo Prazo': 'ARLP',
    'Caixa e Equivalentes de Caixa': 'Caixa',
    'Contas a Receber': 'Contas_Receber',
    'Estoques': 'Estoques',
    'Ativo Total': 'Ativo_Total',
    'Imobilizado': 'Imobilizado',
    'Intangível': 'Intangível',
    }, inplace=True)

# Ordena pelo 'DT_REFER' e 'DENOM_CIA' para garantir a ordem correta
pivot_filtradas.sort_values(by=['DT_REFER', 'DENOM_CIA'], inplace=True)

# Exibe a tabela pivotada
display(pivot_filtradas)

# Salva o DataFrame resultante em um arquivo CSV
pivot_filtradas.to_csv('pivot_filtradas.csv', index=False)
```

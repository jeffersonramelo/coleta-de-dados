```python
import pandas as pd
from langdetect import detect

# Supondo que você já tenha um DataFrame chamado 'df' com a coluna 'texto'
# Você também precisará instalar a biblioteca 'langdetect' se ainda não a tiver

# Função para detectar o idioma de um texto
def detect_language(text):
    try:
        return detect(text)
    except:
        return ''

# Função para remover a parte em português dos textos em inglês
def remove_parte_portugues(text):
    # Verificar se o idioma é inglês
    if detect_language(text) == 'en':
        # Dividir o texto em frases
        sentences = text.split('. ')
        # Manter apenas as frases em inglês
        sentences = [s for s in sentences if detect_language(s) == 'en']
        # Juntar as frases novamente em um texto
        english_text = '. '.join(sentences)
        return english_text
    else:
        return ''

# Aplicar a função detect_language à coluna 'texto' e criar uma nova coluna 'idioma'
df['idioma'] = df['Texto'].apply(detect_language)

# Filtrar apenas as linhas em que o idioma é inglês
df_ingles = df[df['idioma'] == 'en']

# Aplicar a função remove_parte_portugues para remover a parte em português dos textos em inglês
df_ingles['texto_ingles'] = df_ingles['Texto'].apply(remove_parte_portugues)

# Excluir a coluna 'idioma' do novo DataFrame
df_ingles = df_ingles.drop('idioma', axis=1)
```
